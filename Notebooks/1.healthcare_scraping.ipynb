```python
products = soup.find_all("a", class_="css-qlopj4")

print("Total Products:", len(products))
```

```python
for i in range(5):
    print(products[i])
    print("-"*100)
```

```python
product_name = []
price = []
rating = []

for product in products:
    # Product Name
    try:
        name = product.find("h2").text.strip()
    except:
        name = "N/A"

    # Price
    try:
        p = product.find("span", class_="css-111z9ua").text.strip()
    except:
        p = "N/A"

    # Rating
    try:
        r = product.find("div", attrs={"aria-label": True})["aria-label"]
    except:
        r = "N/A"

    product_name.append(name)
    price.append(p)
    rating.append(r)

print("Products extracted:", len(product_name))
```

```python
import pandas as pd

df = pd.DataFrame({
    "Product Name": product_name,
    "Price": price,
    "Rating": rating
})

df.to_csv("healthcare_products.csv", index=False)

print(df.head())
```
