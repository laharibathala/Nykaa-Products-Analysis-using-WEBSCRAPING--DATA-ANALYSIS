```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
links = [p.get_attribute("href") for p in products]

for link in links:
    driver.get(link)

    title = driver.find_element(By.CSS_SELECTOR, "h1").text
    print(title, link)

    driver.back()
```

```python
from bs4 import BeautifulSoup

product_name = []
original_price = []
discounted_price = []
discount = []
reviews = []

for product in products:
    soup = BeautifulSoup(product.get_attribute("outerHTML"), "html.parser")

    # ---------------- Product Name ----------------
    name_tag = soup.find("h2")
    name = name_tag.get_text(" ", strip=True) if name_tag else "Not Available"

    # ---------------- Original Price ----------------
    mrp_tag = soup.select_one("span.css-17x46n5 span")
    op = mrp_tag.get_text(strip=True) if mrp_tag else ""

    # ---------------- Discounted Price ----------------
    sale_tag = soup.select_one("span.css-111z9ua")
    dp = sale_tag.get_text(strip=True) if sale_tag else op

    # If Original Price is blank, use Discounted Price
    if not op:
        op = dp

    # ---------------- Discount ----------------
    disc_tag = soup.select_one("span.css-cjd9an")
    dis = disc_tag.get_text(strip=True) if disc_tag else "No Discount"

    # ---------------- Reviews ----------------
    review_tag = soup.select_one("span.css-1qbvrhp")
    rev = review_tag.get_text(strip=True) if review_tag else "0"

    # Append results
    product_name.append(name)
    original_price.append(op)
    discounted_price.append(dp)
    discount.append(dis)
    reviews.append(rev)
```

```python
df = pd.DataFrame({

    "Product Name": product_name,
    "Original Price": original_price,
    "Discounted Price": discounted_price,
    "Discount": discount,
    "Reviews": reviews

})

# Remove blank product names
df = df[df["Product Name"] != "Not Available"]

# Fill missing Original Price
df["Original Price"] = df["Original Price"].replace("", pd.NA)
df["Original Price"] = df["Original Price"].fillna(df["Discounted Price"])

# Remove duplicate products
df.drop_duplicates(inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)
```
